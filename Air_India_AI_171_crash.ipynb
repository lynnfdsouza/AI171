````python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
t_total = 30  # Total flight time in seconds
t1 = 9.37  # Time to reach maximum altitude in seconds
h_max = 625  # Maximum altitude in feet
x_total = 4921.26  # Total horizontal distance in feet
v_h = x_total / t_total  # Average horizontal speed in ft/s

# Time array
t = np.linspace(0, t_total, 1000)

# Horizontal position
x = v_h * t

# Vertical position
h = np.where(t <= t1, (h_max / t1) * t, h_max - (h_max / (t_total - t1)) * (t - t1))

# Plot flight path
plt.figure(figsize=(10, 6))
plt.plot(x, h, label='Flight Path')
plt.scatter(x[-1], h[-1], color='red', label='Impact Point')
plt.xlabel('Horizontal Distance (ft)')
plt.ylabel('Altitude (ft)')
plt.title('Air India Flight AI 171 Crash Simulation')
plt.legend()
plt.grid(True)
plt.show()

# Impact point
impact_x = x[-1]
impact_h = h[-1]  # Should be close to 0
print(f"Impact Point: x = {impact_x:.2f} ft, h = {impact_h:.2f} ft")

# Debris field radius (based on historical data)
debris_radius = 400 * 3.28084  # 400 meters to feet
print(f"Debris Field Radius: {debris_radius:.2f} ft")
````

The Python code simulates the flight path of Air India Flight AI 171 leading up to a crash. It uses parameters like total flight time, time to reach maximum altitude, maximum altitude, and total horizontal distance to model the flight path.

The simulation calculates and plots the horizontal position (x) and vertical position (h) over time. The output confirms the calculated impact point at the expected horizontal distance and an altitude very close to zero, as well as a calculated debris field radius based on a conversion from meters to feet.

##### Flight Path Simulation
- **Climb Phase (0  t  t1)**:
  - Climb rate: 4000 feet per minute = 66.67 feet per second.
  - Time to 625 feet: t1 = 625 / 66.67  9.37 seconds.
  - Altitude: h(t) = 66.67 * t.
- **Descent Phase (t1 < t  t_total)**:
  - Total time: t_total = 30 seconds.
  - Descent time: t2 = 30 - 9.37 = 20.63 seconds.
  - Descent distance: 625 feet.
  - Average vertical speed: -625 / 20.63  -30.3 feet per second.
  - Assume linear descent: h(t) = 625 - (625 / 20.63) * (t - 9.37).
- **Horizontal Motion**:
  - Total distance: 4921.26 feet.
  - Average speed: v_h = 4921.26 / 30  164.04 feet per second (~97.2 knots).
  - Position: x(t) = 164.04 * t.
